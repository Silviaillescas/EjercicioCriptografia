# Cifrado de Caesar

El Julio César usaba este cifrado para mandar mensajes militares: consiste en desplazar cada letra del alfabeto un número fijo de posiciones.

```python
def caesar_cifrar(mensaje, k):
    resultado = ""
    k = k % 26  # por si k es mayor que 26

    for c in mensaje:
        if c.isalpha():
            base = ord('A') if c.isupper() else ord('a')
            # posición 0-25 -> desplazar -> volver a letra
            resultado += chr((ord(c) - base + k) % 26 + base)
        else:
            resultado += c  # espacios, números, signos se quedan igual

    return resultado


def caesar_descifrar(mensaje_cifrado, k):
    return caesar_cifrar(mensaje_cifrado, -k)


# ---- Prueba rápida ----
if __name__ == "__main__":
    texto = input("Ingrese el mensaje: ")
    k = int(input("Ingrese el desplazamiento (clave): "))

    cifrado = caesar_cifrar(texto, k)
    descifrado = caesar_descifrar(cifrado, k)

    print("\nTexto cifrado:   ", cifrado)
    print("Texto descifrado:", descifrado)
```

¿Por qué lo elegí?

Porque es un cifrado histórico y fácil de entender, ideal para aprender conceptos base: clave, sustitución y cifrado/descifrado.

Ventajas

- Muy fácil de implementar y entender.

- Sirve como introducción a criptografía.

- No cambia caracteres que no sean letras (más cómodo para mensajes reales).

Vulnerabilidades

- Tiene solo 25 claves posibles, así que se rompe rápido por fuerza bruta.

- Es vulnerable a análisis de frecuencia (la letra más repetida suele delatar el idioma).

- No es seguro para uso real hoy; es solo educativo/histórico.
