# conversión de palabras en texto ASCII a BINARIO

```python
def texto_a_binario(texto):
    binario = []
    for caracter in texto:
        binario.append(format(ord(caracter), '08b'))
    return ' '.join(binario)

texto = input("Ingrese el texto a convertir: ")

print("\nTexto ingresado:")
print(texto)

# Conversión a binario
resultado = texto_a_binario(texto)

print("\nTexto en binario:")
print(resultado)
```
