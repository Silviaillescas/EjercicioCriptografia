# conversión de palabras en texto ASCII a BINARIO

```python
tabla_ascii = {
    'A': 65, 'B': 66, 'C': 67, 'D': 68, 'E': 69,
    'F': 70, 'G': 71, 'H': 72, 'I': 73, 'J': 74,
    'K': 75, 'L': 76, 'M': 77, 'N': 78, 'O': 79,
    'P': 80, 'Q': 81, 'R': 82, 'S': 83, 'T': 84,
    'U': 85, 'V': 86, 'W': 87, 'X': 88, 'Y': 89,
    'Z': 90,
    'a': 97, 'b': 98, 'c': 99, 'd': 100, 'e': 101,
    'f': 102, 'g': 103, 'h': 104, 'i': 105, 'j': 106,
    'k': 107, 'l': 108, 'm': 109, 'n': 110, 'o': 111,
    'p': 112, 'q': 113, 'r': 114, 's': 115, 't': 116,
    'u': 117, 'v': 118, 'w': 119, 'x': 120, 'y': 121,
    'z': 122,
    ' ': 32
}

def decimal_a_binario(numero):
    binario = ""
    while numero > 0:
        residuo = numero % 2
        binario = str(residuo) + binario
        numero = numero // 2


    while len(binario) < 8:
        binario = "0" + binario

    return binario

texto = input("Ingrese el texto a convertir: ")

print("\nTexto ingresado:")
print(texto)

print("\nConversión manual a binario:")
for caracter in texto:
    if caracter in tabla_ascii:
        decimal = tabla_ascii[caracter]
        binario = decimal_a_binario(decimal)
        print(f"'{caracter}' → {decimal} → {binario}")
    else:
        print(f"'{caracter}' no está en la tabla ASCII")
```

## Conversión de palabras en texto BASE64 a BINARIO

```python
base64_tabla = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/"

def decimal_a_binario(numero, bits):
    binario = ""
    while numero > 0:
        residuo = numero % 2
        binario = str(residuo) + binario
        numero = numero // 2

    while len(binario) < bits:
        binario = "0" + binario

    return binario


texto_base64 = input("Ingrese el texto en Base64: ")

print("\nTexto Base64 ingresado:")
print(texto_base64)

binario_total = ""

# Conversión Base64 → binario (6 bits)
for caracter in texto_base64:
    if caracter == "=":
        continue  
    valor = base64_tabla.index(caracter)
    binario_6bits = decimal_a_binario(valor, 6)
    binario_total += binario_6bits

# Separar en bytes (8 bits)
print("\nTexto en binario:")
for i in range(0, len(binario_total), 8):
    byte = binario_total[i:i+8]
    if len(byte) == 8:
        print(byte, end=" ")
```

## Conversión de BINARIO a BASE64

```python
base64_tabla = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/"

def binario_a_decimal(bits):
    # Convierte una cadena de bits 
    decimal = 0
    potencia = 1
    for b in bits[::-1]:  # recorre de derecha a izquierda
        if b == '1':
            decimal += potencia
        potencia *= 2
    return decimal

binario = input("Ingrese el binario (8 bits por byte, separado por espacios si quiere): ")

print("\nBinario ingresado:")
print(binario)

# Quita espacios y valida que solo haya 0 y 1
binario = binario.replace(" ", "")

for c in binario:
    if c not in "01":
        print("\nError: el binario solo puede contener 0 y 1 (y espacios).")
        exit()

if len(binario) % 8 != 0:
    print("\nError: el binario debe venir en bytes completos (múltiplo de 8 bits).")
    exit()

bits = binario

faltan = (6 - (len(bits) % 6)) % 6
bits_padded = bits + ("0" * faltan)

# Convertir cada 6 bits a Base64
resultado = ""
for i in range(0, len(bits_padded), 6):
    bloque = bits_padded[i:i+6]
    valor = binario_a_decimal(bloque)
    resultado += base64_tabla[valor]

bits_original = len(bits)
pad_equals = (24 - (bits_original % 24)) % 24
pad_equals = pad_equals // 8 
resultado += "=" * pad_equals

print("\nTexto en Base64:")
print(resultado)
```

## Conversión de BINARIO a ASCII

```python
def binario_a_decimal(bits):
    decimal = 0
    potencia = 1
    
    # Recorrer de derecha a izquierda
    for bit in bits[::-1]:
        if bit == '1':
            decimal += potencia
        potencia *= 2
    
    return decimal


binario = input("Ingrese el binario (8 bits por byte, puede usar espacios): ")

print("\nBinario ingresado:")
print(binario)

# Quitar espacios
binario = binario.replace(" ", "")

# Validar que solo tenga 0 y 1
for c in binario:
    if c not in "01":
        print("\nError: solo se permiten 0 y 1.")
        exit()

# Validar que sea múltiplo de 8
if len(binario) % 8 != 0:
    print("\nError: el binario debe estar en grupos completos de 8 bits.")
    exit()

texto = ""

# Procesar cada byte
for i in range(0, len(binario), 8):
    byte = binario[i:i+8]
    decimal = binario_a_decimal(byte)
    caracter = chr(decimal)  # Convertir decimal a carácter ASCII
    texto += caracter

print("\nTexto en ASCII:")
print(texto)
```
